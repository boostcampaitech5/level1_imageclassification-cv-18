```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
submission_df = pd.read_csv("./csvs/output.csv")
```

# Hard Voting 

```python
preds = []
for i in os.listdir("./"):
    if i.endswith("csv"):
        print(i)
        last = pd.read_csv(i)
        mask_preds = np.array(last.iloc[:, :3])
        gender_preds = np.array(last.iloc[:, 3:5])
        age_preds = np.array(last.iloc[:, 5:])
        result = np.argmax(mask_preds, axis=1)*6
        result = result + np.argmax(gender_preds, axis=1)*3
        result = result + np.argmax(age_preds, axis=1)

        preds.append(result)
```

```python
hard_votes = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=0, arr=preds)

# Print the results
print("Hard votes along each column:")
print(hard_votes)
```

```python
submission_df["ans"] = hard_votes
```

```python
fig,axes = plt.subplots(2, 3, figsize=(25, 12))
axes = axes.flatten()
for i in range(5):
    pd.Series(preds[i]).value_counts().sort_index().plot(kind="bar",ylim=[0,2500], ax=axes[i])
pd.Series(submission_df["ans"]).value_counts().sort_index().plot(kind="bar",ylim=[0,2500], ax=axes[5])


# pd.Series(preds[0]).value_counts().sort_index().plot(kind="bar", figsize=(14, 10),ylim=[0,2500])
```

```python
submission_df.to_csv("Final_2.csv")
```

# Soft voting 

```python
submissions = []
for i in os.listdir("./"):
    if i .endswith("csv"):
        print(i)
        last = pd.read_csv(i)
        submissions.append(last)
```

```python
final = submissions[0] + submissions[1] +submissions[2] +submissions[3] +submissions[4] 
```

```python
mask_preds = np.array(final.iloc[:, :3])
gender_preds = np.array(final.iloc[:, 3:5])
age_preds = np.array(final.iloc[:, 5:])
```

```python
mask_res = np.argmax(mask_preds, axis=1)
gender_res = np.argmax(gender_preds, axis=1)
age_res = np.argmax(age_preds, axis=1)
```

```python
ans = mask_res*6 + gender_res*3 + age_res
```

```python
submission_df["ans"] = ans
```

```python
fig,axes = plt.subplots(2, 3, figsize=(25, 12))
axes = axes.flatten()
for i in range(5):
    pd.Series(preds[i]).value_counts().sort_index().plot(kind="bar",ylim=[0,2500], ax=axes[i])
pd.Series(submission_df["ans"]).value_counts().sort_index().plot(kind="bar",ylim=[0,2500], ax=axes[5])


# pd.Series(preds[0]).value_counts().sort_index().plot(kind="bar", figsize=(14, 10),ylim=[0,2500])
```

```python
submission_df.to_csv("Final_1.csv")
```
